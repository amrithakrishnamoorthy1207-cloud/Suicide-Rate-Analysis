## Data Cleaning and Preparation

The original dataset contains several metadata columns that are not required for this analysis.

To simplify the dataset and focus only on relevant variables, we will extract the following columns:

- Year
- Country
- Sex
- Age group
- Suicide rate per 100,000 population

This cleaned dataset will be used for further analysis and visualization.

```python
# Select only useful columns

df_clean = df[['DIM_TIME',
               'GEO_NAME_SHORT',
               'DIM_SEX',
               'DIM_AGE',
               'RATE_PER_100000_N']].copy()

# Rename columns to simpler names

df_clean.columns = ['Year','Country','Sex','Age','Suicide_Rate']

# Show first rows of cleaned dataset
df_clean.head()
```

## Filtering Total Population Data

The dataset contains separate records for males, females, and the total population.

To simplify the analysis and avoid duplicate entries, we filter the dataset to include only records where:

- Sex = TOTAL
- Age = TOTAL

This ensures that each country has a single suicide rate value per year.

```python
# Filter dataset for total population

df_total = df_clean[(df_clean['Sex'] == 'TOTAL') & (df_clean['Age'] == 'TOTAL')]

# Display first rows
df_total.head()
```

## Global Suicide Rate Trend (2000–2024)

To understand how suicide rates have changed over time, we analyze the global average suicide rate across years.

A line chart is used to visualize the trend of suicide mortality rates from 2000 to the most recent available year in the dataset.

This visualization helps identify whether suicide rates are increasing, decreasing, or remaining stable over time.

```python
# Extract global data (World)

global_data = df_total[df_total['Country'] == 'World']

# Plot global suicide rate trend

plt.figure(figsize=(10,6))

plt.plot(global_data['Year'], global_data['Suicide_Rate'], marker='o')

plt.title("Global Suicide Rate Trend (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Suicide Rate per 100,000 population")

plt.show()
```

## Interpretation of the Global Trend

The line chart shows the global suicide mortality rate per 100,000 population from 2000 onwards.

From the visualization, it can be observed that the global suicide rate has gradually declined over the years. The rate was above 12 deaths per 100,000 population in the early 2000s and decreased to around 9 deaths per 100,000 population in recent years.

This downward trend may reflect improvements in mental health awareness, prevention strategies, and healthcare interventions implemented in many countries.

### Peak Suicide Age Group in India for 2019

This chart shows the suicide rate across different age groups in India for the year 2019.

Since age-wise data in the dataset is available only for 2019, the analysis identifies which age group had the highest suicide rate during that year. This helps highlight the most affected age category based on the available age-specific data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset if not already loaded
df = pd.read_csv("/content/16BBF41_ALL_LATEST.csv")

# Filter India, 2019, TOTAL sex, and age groups only
india_2019_age = df[
    (df["GEO_NAME_SHORT"] == "India") &
    (df["DIM_TIME"] == 2019) &
    (df["DIM_SEX"] == "TOTAL") &
    (df["DIM_AGE"] != "TOTAL")
].copy()

# Convert rate column
india_2019_age["RATE_PER_100000_N"] = pd.to_numeric(india_2019_age["RATE_PER_100000_N"], errors="coerce")

# Age label mapping
age_map = {
    "Y15T19": "15-19",
    "Y15T24": "15-24",
    "Y15T29": "15-29",
    "Y25T34": "25-34",
    "Y30T49": "30-49",
    "Y35T44": "35-44",
    "Y45T54": "45-54",
    "Y55T64": "55-64",
    "Y65T74": "65-74",
    "Y75T84": "75-84",
    "Y_GE85": "85+"
}

india_2019_age["Age_Label"] = india_2019_age["DIM_AGE"].map(age_map)

# Age order
age_order = ["15-19", "15-24", "15-29", "25-34", "30-49", "35-44", "45-54", "55-64", "65-74", "75-84", "85+"]
india_2019_age["Age_Label"] = pd.Categorical(india_2019_age["Age_Label"], categories=age_order, ordered=True)
india_2019_age = india_2019_age.sort_values("Age_Label")

# Find peak age group
peak_row = india_2019_age.loc[india_2019_age["RATE_PER_100000_N"].idxmax()]
print("Peak age group in 2019:", peak_row["Age_Label"])
print("Highest suicide rate:", peak_row["RATE_PER_100000_N"])

# Plot
plt.figure(figsize=(12,6))
bars = plt.bar(india_2019_age["Age_Label"], india_2019_age["RATE_PER_100000_N"])

# Highlight peak bar by annotation
for bar, age, rate in zip(bars, india_2019_age["Age_Label"], india_2019_age["RATE_PER_100000_N"]):
    plt.text(bar.get_x() + bar.get_width()/2, rate + 0.4, f"{rate:.2f}", ha="center", fontsize=8)

plt.title("Suicide Rate by Age Group in India (2019)")
plt.xlabel("Age Group")
plt.ylabel("Suicide Rate per 100,000")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
